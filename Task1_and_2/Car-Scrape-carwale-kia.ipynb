## Methodology of Data Collection:

```python

urlsite = "https://www.carwale.com/kia-cars/seltos/userreviews"
```

## Algorithm of scraping
All the reviews from carwale.com for cars Kia Seltos(as an example shown here) are available in "https://www.carwale.com/kia-cars/seltos/userreviews-p<page-no>" where the page no ranges between 1-21. Hence we have scrolled all the 21 pages with appropriate urls.
    
Each page consists of mutliple user reviews in brief. To get the full data, per review, the following methodology is followed:
-The URL which details the review is captured from the main page
-A new window is visited with the specific url(which details the reviews)
-All possible data about in the review captured
-A dictionary with keys as 'revtext', 'date', 'rating', 'site' is filled
-A row is added to the CSV file per review
-The new window is closed.


```python
try:
    for i in range(1,21,1):
        urlmain = urlsite+'-p'+str(i)+'/'
        print("urlmain",urlmain)
        driver.get(urlmain)
        childpages = driver.find_elements_by_xpath("//a[@data-cwtcact='TitleClick']")

        for pageref in childpages:

            page = pageref.get_attribute('href')
            id = page.split('/')[-2]
            print("id=",id)
            #get the id here
            review_dict = {}  
            #*******************************************************
            # Open a new window
            driver.execute_script("window.open('');")
            # Switch to the new window and open URL B
            driver.switch_to.window(driver.window_handles[1])   
            #***********************************************
            url=urlsite+'/'+str(id)
            print("Url is:"+url)
            driver.get(url)
            time.sleep(5)
            reviewlist = driver.find_elements_by_xpath("//div[@categoryname='userreviews']")
            print(len(reviewlist))
            for review in reviewlist:

                # get the review text
                try:
                    #revtext = review.find_element_by_xpath(".//div[@class='contentspace']//a") #For just the heading
                    revtext = review.find_elements_by_xpath(".//div[@class='mid-box margin-top20']")[3]    # For title and heading
                    reviewtext = revtext.text.split('\n')[1]
                except:
                    reviewtext = ''
                # get the review date    
                try:
                    date = review.find_elements_by_xpath(".//h4")[0]    # For title and heading
                    reviewdate = date.text.split("on")[2].lstrip().rstrip()
                    dates = reviewdate.split(',')
                    fdate = dates[0].split(' ')
                    reviewdate = fdate[1]+' '+fdate[0]+', '+ dates[1].lstrip().rstrip()
                except:
                    reviewdate = ''
                # get the review ratings
                try:
                    rating =0 
                    reviewrating = review.find_elements_by_xpath(".//ul[@class='model-ratings vr-dotted']//li")[0]    # For title and heading
                    reviewrating = reviewrating.find_elements_by_xpath(".//img[@class='text-bottom']")
                    print(len(reviewrating))
                    for ratings in reviewrating:
                        if("1.png" in ratings.get_attribute('src')):
                            rating = rating+1

                    #reviewrating = reviewrating.text
                except:
                    rating = 0

                review_dict['revtext']    = reviewtext
                review_dict['date']    = reviewdate
                review_dict['rating'] = str(rating)
                review_dict['site'] = "carwale.com"
            # Close the tab with URL B
            driver.close()
            # Switch back to the first tab with URL A
            driver.switch_to.window(driver.window_handles[0])
            #*********************************************************
            print(review_dict)
            writer.writerow(review_dict.values())
except Exception as e:
    print(e)
    csv_file.close()
```
